```python
class problemEight():
    """ Description: This class prints the kmers in the sequence """
    """ Input: The input is the length of the kmer and the sequence """
    """ Ouput: The output of this method is the kmers """
    def __init__ (self, kmerLength, sequence):
        """ Description: This method initializes variables to be used within the class """
        """ Input: The input is the length of the kmer and the sequence """
        """ Ouput: The output of this method is nothing, but these variables are initialized as self"""
        self.kmerLength = kmerLength #Initializes the length of the kmer as self
        self.sequence = sequence #Initializes the sequence as self
        self.totalLength = len(sequence) #Initializes the total length of the sequence as self

    def printKmers (self):
        """ Description: The print kmers method prints all the kmers in the sequence """
        """ Input: The method does not have any inputs but uses in the previously initialized values"""
        """ Output: The method prints out the kmers in the sequences in a sorted fashion """
        kmerList = []
        for i in range(int(self.totalLength)-(int(self.kmerLength))+1): #Goes in from the beginning to the end-kmerlength so that all valid kmers are printed
            kmerList.append(self.sequence[i:i+(int(self.kmerLength))]) #Kmer is appended to a list of kmers
        for i in (sorted(kmerList)): #Sorts the kmerList and prints it out one by one
            print(i)

def main(inFile):
    """ Description: The main reads in the input file, and outputs all the kmers in the sequence with a given kmer length"""
    """ Input: The main reads in an input file containing a sequences and kmer length """
    """ Output: The main runs the functions/methods required to print the kmers """
    with open(inFile, "r") as file:
        length = file.readline().rstrip() #Removes the newline character
        sequence = file.readline().rstrip() #Removes the newline character
    val = problemEight(length, sequence) #Initializes certain variables
    val.printKmers() #Indentifies and prints out the kmers in the sequence

if __name__== "__main__":
    main("rosalind_ba3a.txt")
```
