```python
class problemThirteen():
    """ Description: This class identifies the eularian path """
    """ Input: The input contains the different pathways """
    """ Ouput: The output is a formatted path list"""
    def __init__(self, graph):
        """ Description: This method initializes the variables """
        """ Input: The input is a dictionary containing the start nodes and corresponding end nodes """
        """ Output: There is no output, but the graph is initialized """
        self.graph = graph

    def joinElem(self, l):
        """ Description: This method formats the output by seperating by arrows """
        """ Input: The input is path list """
        """ Ouput: The output is a formatted path list"""
        out = '' #Initializes as string
        for element in l: #Goes through the list
            out += str(element) + ("->") #Seperates using arrows
        return out[:-len("->")] #Returns the path formatted in the correct manner

    def eulerPath(self, vertices):
        """ Description: Determines the eularian path given the vertices """
        """ Input: Recieves vertices from main """
        """ Ouput: The output is a eularian path list"""
        ins = {}; outs = {}
        for i in vertices: #Creates a dictionary for ins of a node
            ins[i] = 0 #Sets each to 0
        for j in vertices: #Creates a dictionary for outs of a node
            outs[j] = 0 #Sets each to 0

        for vertex in vertices: #Iterates through the vertices
            if vertex in self.graph: #Checks if vertex is in the graph dictionary
                outs[vertex] = len(self.graph[vertex]) #Sets to the end nodes of the start
            for adj in self.graph[vertex]: #Iterates through the ends of that start
                ins[adj] += 1 #Adds to the count of adjacent

        for vertex in vertices: #Iterates through the vertices
            if ins[vertex] < outs[vertex]: #Checks to see which vertext is the start node
                startNode = vertex #Sets the start node to the vertex

        currentPath = [startNode] #Begins from the start node
        path = []
        while len(currentPath) > 0: #Goes to the length of the current path
            if outs[startNode]:
                currentPath.append(startNode) #Appends to the path
                nextVertex = self.graph[startNode].pop() #Identifies the next vertex
                outs[startNode] -= 1 #Removes from the count of outs
                startNode = nextVertex #Moves onto the next vertex
            else:
                path.append(startNode) #Adds to the path
                startNode = currentPath.pop() #Removes from current path
        path.reverse() #Flips the elements of the list
        return path #Returns the final pathway

def main(inFile):
    """ Description: This method recieves an input containing the different pathways """
    """ Input: The input contains the different pathways """
    """ Output: The output is a formatted path list"""
    graph = {}
    vertices = set()
    with open(inFile, 'r') as f:
        for line in f: #Iterates through the lines of the file
            lineList = ','.join(line.split(" ")).split(",") #Determines a list of character in each line
            start = int(lineList[0]) #Identifies the start node
            ends = [int(n) for n in lineList[2:]] #Identifies the end nodes
            vertices.add(start) #Adds the start node to the set of vertices
            #vertices = vertices #| set(ends) #Vertices is set equal to one of these
            graph[start] = ends #Adds start and ends nodes to dictionary
        processed = problemThirteen(graph) #Initializes with the graph dictionary
        path = processed.eulerPath(list(vertices)) #Sends in the vertices to the eulerPath function
        print((processed.joinElem(path))) #Prints out the eularian path in the appropriate manner

if __name__ == "__main__":
    main('rosalind_ba3g.txt')
```
